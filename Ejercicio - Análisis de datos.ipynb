Simule 10000 escenarios para el siguiente año y calcule la probabilidad de que los contagios sean menores que 50.

```python
def cont_sim(media,std,inicio,dias,N):
    c=  np.random.normal(media,std,size=(dias,N))
    fechas = pd.date_range(inicio, periods=dias)
    return pd.DataFrame(index = fechas, data = c)
    
    
```

```python
cont_sim(0,1,"2021-08-06",365,10000).cumprod()*8
```
